### Paso 2: Crear función para mandar texto al endpoint de ex-human y regresar video.

Para más información: https://docs.exh.ai/reference/generate_lipsync_animations_v2_generate_lipsync_post

```python
def generate_video(texto: str, key: str) -> str:
    """
    Convierte Texto a Video Bytes.
    """
    # Payload
    body = {
      'bot_name': 'Your-bot-name', # Cambiar por el nombre de tu bot
      'bot_response': texto,
      'voice_name': "Fay",
      'animation_pipeline': "high_speed",
      'idle_url': "your-idle-url" # Cambiar por tu idle url
    }
    
    # Mandando el request para convertir texto a video.
    endpoint = "https://api.exh.ai/animations/v1/generate_lipsync"
    headers = {"Authorization": f"Bearer {key}"}
    res = requests.post(endpoint, json=body, headers=headers)
    
    return res.content
```

### Paso 4: Guardando los video bytes dentro de un archivo de video.mp4

```python
def save_video(video_bytes: str, filename: str) -> None:
    """
    Guarda video bytes dentro de archivo de video.
    """
    with open(filename, "wb") as fh:
      fh.write(video_bytes)
```

### Paso 6: Generar llamada a servidores de OpenAI (Recapitulando Demo Az Bot)

En la siguiente sección se reusarán las funciones que fueron creadas en talleres anteriores para obtener respuestas en bases a distintas preguntas.

```python
import openai, time
```

```python
openai.api_type = "azure"
openai.api_base = "your-endpoint" # cambiar por tu endpoint
openai.api_version = "2023-03-15-preview"
openai.api_key = "Your-azure-api-key" # Cambiar por tu azure api key
```

```python
def get_history(previous_questions: list, previous_answers: list) -> str:
    """
    Obtiene el historial de la conversación.
    """
    history = []
    for (question, response) in zip(previous_questions, previous_answers):
        quest = {"role": "user", "content": f'{question}'}
        ans = {"role": "assistant", "content": f'{response}'}
        history.append(quest)
        history.append(ans)
    return history
```

```python
def ask_gpt(question: str, history: str) -> str:
    """
    Obtiene respuesta en base a pregunta y contezto
    """
    quest=  {"role": "user", "content": f'{question}'}

    model_engine = "gpt-35-turbo" # Cambiar a tu deployment name ********
    
    messages = [
      {"role": "system", "content": "Your name is America. You are a helpful virtual assistant. You can answer any question. Respond to answers in a single line. Answers should be as short as possible"},
      ]
 
    prompt = messages + history
    prompt.append(quest)

    response = openai.ChatCompletion.create(
      engine=model_engine,
      messages=prompt,
      max_tokens=600,
      temperature=0.9,
      stream=True
      )
    answer = ''
    full_response = ''

    for event in response: 
        print(answer, end='', flush=True)
        event_text = event['choices'][0]['delta']
        answer = event_text.get('content', '')
        full_response += answer

    return full_response
```

```python
def display_video(video_path: str, loop: bool = False):
    video = open(video_path, "rb").read()
    video_widget = Video(value=video, width=240, height=160, autoplay=True, loop=loop, end=180, controls=False)
    display(video_widget)
```

### Paso 7: Integrando el avatar y el bot

```python
previous_questions = []
previous_answers = []

#display_video('IMSS_ejemplo.mp4')

counter = 0
while True:
    if counter == 0:
        display_video('idle.mp4')
    else:
        display_video('test.mp4')
        display_video('idle.mp4')
    user_input = input("Pregunta:")
    print()
    history = get_history(previous_questions, previous_answers)
    answer = ask_gpt(user_input, history)
    video_content = generate_video(answer, API_KEY)
    save_video(video_content, 'test.mp4')
    previous_questions.append(user_input)
    previous_answers.append(answer)
    counter += 1
```
